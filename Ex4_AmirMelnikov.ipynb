# Data Visualization

In this section, you should **ignore** the columns RowNumber, CustomerId, and Surname. They will not be a part of the analysis.

The next exercises are similar to the Boston Housing example that was shown in class.

**Exercise 7** (5 points): Create a list with the names of the continuous variables and a list with the names of the discrete variables. Exercise 5 should help you identify these variables. You may assume that a continuous variable has more than 10 unique possible values.

```python
discrete, cont = [], []
for col in df.columns:
  if df[col].nunique() <= 10:
    discrete.append(col)
  else:
    cont.append(col)
    
print(discrete)
print(cont)
```

**Exercise 8** (10 points): Plot the distribution of all the continuous variables using **histograms**.

```python
fig = plt.figure(figsize=(18, 10))
for i in range(len(cont)):
  fig.add_subplot(4, 3, i+1)
  sns.histplot(df[cont[i]])
plt.tight_layout()
plt.show()
```

**Exercise 9** (10 points): Use **countplots** to visualize the distributions of the discrete columns.

```python
fig = plt.figure(figsize=(14, 12))
for i in range(len(discrete)):
  fig.add_subplot(4, 3, i+1)
  sns.countplot(x=df[discrete[i]])
plt.tight_layout()
plt.show()
```

The dependent variable of the analysis is "Exited" - i.e., whether a customer left the bank or not.

**Exercise 10** (15 points): Compare the distribution of each continuous variable for past customers and existing customers. For each column, plot both distributions on the same figure using **kdeplots**.

```python
fig, ax = plt.subplots(1, figsize=(6, 4))
sns.kdeplot(df[df["Exited"] == 0]["Age"], fill=True, ax=ax, label="Bank costumer")
sns.kdeplot(df[df["Exited"] == 1]["Age"], fill=True, ax=ax, label="Left the bank")
ax.legend()

fig, ax = plt.subplots(1, figsize=(6, 4))
sns.kdeplot(df[df["Exited"] == 0]["CustomerId"], fill=True, ax=ax, label="Bank costumer")
sns.kdeplot(df[df["Exited"] == 1]["CustomerId"], fill=True, ax=ax, label="Left the bank")
ax.legend()

fig, ax = plt.subplots(1, figsize=(6, 4))
sns.kdeplot(df[df["Exited"] == 0]["CustomerId"], fill=True, ax=ax, label="Bank costumer")
sns.kdeplot(df[df["Exited"] == 1]["CustomerId"], fill=True, ax=ax, label="Left the bank")
ax.legend()

fig, ax = plt.subplots(1, figsize=(6, 4))
sns.kdeplot(df[df["Exited"] == 0]["CreditScore"], fill=True, ax=ax, label="Bank costumer")
sns.kdeplot(df[df["Exited"] == 1]["CreditScore"], fill=True, ax=ax, label="Left the bank")
ax.legend()

fig, ax = plt.subplots(1, figsize=(6, 4))
sns.kdeplot(df[df["Exited"] == 0]["Tenure"], fill=True, ax=ax, label="Bank costumer")
sns.kdeplot(df[df["Exited"] == 1]["Tenure"], fill=True, ax=ax, label="Left the bank")
ax.legend()

fig, ax = plt.subplots(1, figsize=(6, 4))
sns.kdeplot(df[df["Exited"] == 0]["Balance"], fill=True, ax=ax, label="Bank costumer")
sns.kdeplot(df[df["Exited"] == 1]["Balance"], fill=True, ax=ax, label="Left the bank")
ax.legend()
```
